# Encdoing categorical features

```python
print(train.shape, test.shape)
target = train['target']
train.drop('target', axis=1, inplace=True)
train = pd.get_dummies(train)
test = pd.get_dummies(test)
print(train.shape, test.shape)
```

```python
train.head()
```

# Modelling

```python
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
```

```python
# Decision Tree
dt = DecisionTreeRegressor(splitter='best', max_depth=8)
dt.fit(X_train, y_train)
print("Decision Tree on train")
y_pred_dt_train = dt.predict(X_train)
print("RMSE on train (Decision Tree):", np.sqrt(mean_squared_error(y_pred_dt_train, y_train)))
print("Decision Tree on validation")
y_pred_dt_val = dt.predict(X_val)
print("RMSE on validation (Decision Tree):", np.sqrt(mean_squared_error(y_pred_dt_val, y_val)))
```

```python
# Decision Tree on train
# RMSE on train (Decision Tree): 0.8601209444537481
# Decision Tree on validation
# RMSE on validation (Decision Tree): 0.8700752332944751
```

```python
rf = RandomForestRegressor(n_estimators=30, max_depth=20)
rf.fit(X_train, y_train)
print("Random Forest on train")
y_pred_rf_train = rf.predict(X_train)
print("RMSE on train (Random Forest):", np.sqrt(mean_squared_error(y_pred_rf_train, y_train)))
print("Decision Tree on validation")
y_pred_rf_val = rf.predict(X_val)
print("RMSE on validation (Random Forest):", np.sqrt(mean_squared_error(y_pred_rf_val, y_val)))
```

```python
# Random Forest on train
# RMSE on train (Random Forest): 0.7733943645328599
# Decision Tree on validation
# RMSE on validation (Random Forest): 0.860031503342137
```
